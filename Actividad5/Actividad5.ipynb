```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Leer archivo de datos de 00Z
# Convertir la columna CAPE de objeto a número
df00 = pd.read_csv("df2017CAPE_PW_00Z.csv", header=None, names=['Date', 'CAPE', 'PW'])
df00.CAPE=pd.to_numeric(df00.CAPE, errors='coerce')
df00.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate' (00Z)
df00['Ndate'] = pd.to_datetime(df00['Date'], format='%d %m %Y')
df00['month'] = df00['Ndate'].dt.month
df00.head()
```

```python
# Leer archivo de datos de 12Z
# Convertir la columna CAPE de objeto a número
df12 = pd.read_csv("df2017CAPE_PW_12Z.csv", header=None, names=['Date', 'CAPE', 'PW'])
df12.CAPE=pd.to_numeric(df12.CAPE, errors='coerce')
df12.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'(12Z)
df12['Ndate'] = pd.to_datetime(df12['Date'], format='%d %m %Y')
df12['month'] = df12['Ndate'].dt.month
df12.head()
```

```python
# graficar Boxplots por mes de CAPE (00Z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="CAPE", data=df00)
plt.show()
```

```python
# graficar Boxplots por mes de PW (00Z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="PW", data=df00)
plt.show()
```

```python
# graficar Boxplots por mes de CAPE (12Z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="CAPE", data=df12)
plt.show()
```

```python
# graficar Boxplots por mes de PW (12Z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="PW", data=df12)
plt.show()
```

```python
#Gráfica jointplot (00Z)
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("CAPE", "PW", data=df00, kind="reg",
                   color="r", size=7)
plt.show(g)
```

```python
#Gráfica jointplot (12Z)
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("CAPE", "PW", data=df12, kind="reg",
                   color="r", size=7)
plt.show(g)
```

```python
#Gráfica lmplot (00Z)
g = sns.lmplot(x="CAPE", y="PW", hue="month",
               truncate=True, size=5, data=df00)
plt.show(g)
```

```python
#Gráfica lmplot (12Z)
g = sns.lmplot(x="CAPE", y="PW", hue="month",
               truncate=True, size=5, data=df12)
plt.show(g)
```
