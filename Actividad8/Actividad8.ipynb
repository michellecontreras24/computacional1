```python
from scipy import array

def VanDerPol (X,t=0):
    x=X[0]
    y=X[1]
    a=X[0]
    b=X[1]
    c=X[0]
    d=X[1]
    e=X[0]
    f=X[1]
    g=X[0]
    h=X[1]
    i=X[0]
    j=X[1]
    k=X[0]
    l=X[1]
    m=X[0]
    n=X[1]
    o=X[0]
    p=X[1]
    q=X[0]
    r=X[1]
    dx=X[1]
    dy= u*(1-X[0]*X[0])*X[1]-X[0]
    return array ([dx,dy])
```

```python
#Gráfica 1
#Phase portrait of the unforced Van der Pol oscillator, showing a limit cycle and the direction field

import numpy as np
from scipy import arange
from numpy import loadtxt
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from pylab import figure, plot, xlabel, grid, hold, legend, title, savefig, ylabel
from matplotlib.font_manager import FontProperties
import pylab as p


#Tiempo
t0=0
tmax=50
pastemps=0.01
t=arange(t0,tmax,pastemps)

#Mu
u=2.0

#Condiciones iniciales 1
x0=-3.0
y0=-3.0
#Solución de la ED
x,y=odeint(VanDerPol, (x0,y0),t).T

#Condiciones iniciales 2
a0=-2.0
b0=4.0

#Solución de la ED
a,b=odeint(VanDerPol, (a0,b0),t).T

#Condiciones iniciales 3
c0=-1.0
d0=-2.0

#Solución de la ED
c,d=odeint(VanDerPol, (c0,d0),t).T

#Condiciones iniciales 4
e0=1.0
f0=2.0

#Solución de la ED
e,f=odeint(VanDerPol, (e0,f0),t).T

#Condiciones iniciales 5
g0=-2.0
h0=0.0

#Solución de la ED
% matplotlib inline
g,h=odeint(VanDerPol, (g0,h0),t).T

#Vectores
nb_points   = 20

i = np.linspace(-6,6, nb_points)
j = np.linspace(-6,6, nb_points)

I1 , J1  = np.meshgrid(i, j)                       
DI1, DJ1 = VanDerPol([I1, J1])                    
M = (np.hypot(DI1, DJ1))
M[ M == 0] = 1.          
DI1 /= M                               
DJ1 /= M

Q = p.quiver(I1, J1, DI1, DJ1, M, pivot='mid', cmap=p.cm.bone)

#Gráfica 1
figure(1, figsize=(10, 6))

xlabel('x', fontsize=15)
ylabel('y', fontsize=15)
grid(True)
lw = 2
plt.ylim(-6,6)
plt.xlim (-6,6)

plot(x,y, 'b', linewidth=lw)
plot(a,b, 'b', linewidth=lw)
plot(c,d, 'b', linewidth=lw)
plot(e,f, 'b', linewidth=lw)
plot(g,h, 'r', linewidth=lw)

title('Phase Portrait of the unforced Van der Pol oscillator', fontsize=15)
savefig('Graf1.png', dpi=100)
```

```python
def VanDerPol (X,t):
    x=X[0]
    y=X[1]
    dx=X[1]
    dy= u*(1-X[0]*X[0])*X[1]-X[0]+a*np.sin(w*t)
    return array ([dx,dy])
```

```python
#Gráfica 4
#Chaotic behaviour in the Van der Pol oscillator with sinusoidal forcing. 

#Tiempo
t0=300
tmax=600
pastemps=0.01
t=arange(t0,tmax,pastemps)

#Condiciones iniciales 
x0=-1.0
y0=0.0

#Mu 
u=8.53

#Amplitud de la fuerza
a=1.2

#Frecuencia de la fuerza
w=2.0*np.pi/10.0

#Solución de la ED
x,y=odeint(VanDerPol, (x0,y0),t).T

#Gráfica 4
figure(4, figsize=(15, 3.5))

xlabel('t', fontsize=10)
ylabel('$x_1$', fontsize=10)
grid(True)
lw = 1
plt.ylim(-2.5,2.5)
plt.xlim (300,600)

plot(t,x, 'b', linewidth=lw)

title('Behaviour in the Van der Pol oscillator with sinusoidal forcing', fontsize=20)
savefig('Graf4.png', dpi=100)

```
